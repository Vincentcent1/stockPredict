```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("../mined/combine/exxonProcessed.csv", index_col=False)
df = df.drop(df.columns[0],axis=1)
```

```python
df["Year"] = df["StartDate"]%10000
df["Date"] = (df["StartDate"]%1000000/10000).astype(int)
df["Month"] = (df["StartDate"]/1000000).astype(int)
```

```python
df = df.sort_values(by=["Year", "Month", "Date"])
df = df.loc[(df["error"] != "Yes")]
```

```python
df
```

```python
df2 = pd.read_csv("../processedSelected/xom.csv", index_col = False)
```

```python
df2["StockScore"] = (df2["Delta"]/df2["Open"])/0.2
```

```python
df2["reuter_score"] = 0
df2["reuter_credibility"] = 0
df2["reuter_cumu"] = 0
df2['fool_score'] = 0
df2['fool_credibility'] = 0
df2["fool_cumu"] = 0
df2["invest_score"] = 0
df2["invest_credibility"] = 0
df2["invest_cumu"] = 0
df2["cnbc_score"] = 0
df2["cnbc_credibility"] = 0
df2["cnbc_cumu"] = 0
df2["market_score"] = 0
df2["market_credibility"] = 0
df2["market_cumu"] = 0
```

```python
df2
```

```python
for index, row in df2.iterrows():
    year = int(row["Date"].split("-")[0])
    month = int(row["Date"].split("-")[1])
    date = int(row["Date"].split("-")[2])
    reuter_sentiment_score = 0
    reuter_sentiment_count = 0
    fool_sentiment_score = 0
    fool_sentiment_count = 0
    invest_sentiment_score = 0
    invest_sentiment_count = 0
    cnbc_sentiment_score = 0
    cnbc_sentiment_count = 0
    market_sentiment_score = 0
    market_sentiment_count = 0
    reading_frame = df.loc[(df["Year"] == year) & (df["Month"] == month) & (df["Date"] == date)]
    if not reading_frame.empty:
        for reading_index, reading_row in reading_frame.iterrows():
            if reading_row["Site"] == "Reuters":
                reuter_sentiment_score += reading_row["sv_score"]
                reuter_sentiment_count += 1
            elif reading_row["Site"] == "Investopedia":
                invest_sentiment_score += reading_row["sv_score"]
                invest_sentiment_count += 1
            elif reading_row['Site'] == "CNBC":
                cnbc_sentiment_score += reading_row["sv_score"]
                cnbc_sentiment_count += 1
            elif reading_row['Site'] == "Fool":
                fool_sentiment_score += reading_row['sv_score']
                fool_sentiment_count += 1
            elif reading_row['Site'] == 'MarketWatch':
                market_sentiment_score += reading_row["sv_score"]
                market_sentiment_count += 1
    if reuter_sentiment_count > 0:
        df2.loc[index,"reuter_score"] = reuter_sentiment_score/reuter_sentiment_count
        df2.loc[index,"reuter_credibility"] = 1-(abs(row["reuter_score"] - row["StockScore"]))/2
        if index == 0:
            df2.loc[index,"reuter_cumu"] = 1-(abs(row["reuter_score"] - row["StockScore"]))/2
        else:
            if not pd.isna(df2.loc[index-1, "reuter_cumu"]):
                df2.loc[index, "reuter_cumu"] = (df2.loc[index, "reuter_credibility"]+df2.loc[index-1, "reuter_cumu"])/2
            else:
                df2.loc[index, "reuter_cumu"] = 1-(abs(row["reuter_score"] - row["StockScore"]))/2            
    else:
        df2.loc[index,"reuter_score"] = None
        df2.loc[index,"reuter_credibility"] = None
        if index == 0:
            df2.loc[index, "reuter_cumu"] = None
        else:
            if not pd.isna(df2.loc[index-1, "reuter_cumu"]):
                df2.loc[index, "reuter_cumu"] = df2.loc[index-1, "reuter_cumu"]
            else:
                df2.loc[index, "reuter_cumu"] = None
                
    if fool_sentiment_count > 0:
        df2.loc[index,"fool_score"] = fool_sentiment_score/fool_sentiment_count
        df2.loc[index,"fool_credibility"] = 1-(abs(row["fool_score"] - row["StockScore"]))/2
        if index == 0:
            df2.loc[index,"fool_cumu"] = 1-(abs(row["fool_score"] - row["StockScore"]))/2
        else:
            if not pd.isna(df2.loc[index-1, "fool_cumu"]):
                df2.loc[index, "fool_cumu"] = (df2.loc[index, "fool_credibility"]+df2.loc[index-1, "fool_cumu"])/2
            else:
                df2.loc[index, "fool_cumu"] = 1-(abs(row["fool_score"] - row["StockScore"]))/2            
    else:
        df2.loc[index,"fool_score"] = None
        df2.loc[index,"fool_credibility"] = None
        if index == 0:
            df2.loc[index, "fool_cumu"] = None
        else:
            if not pd.isna(df2.loc[index-1, "fool_cumu"]):
                df2.loc[index, "fool_cumu"] = df2.loc[index-1, "fool_cumu"]
            else:
                df2.loc[index, "fool_cumu"] = None
        
    if invest_sentiment_count > 0:
        df2.loc[index,"invest_score"] = invest_sentiment_score/invest_sentiment_count
        df2.loc[index,"invest_credibility"] = 1-(abs(row["invest_score"] - row["StockScore"]))/2
        if index == 0:
            df2.loc[index,"invest_cumu"] = 1-(abs(row["invest_score"] - row["StockScore"]))/2
        else:
            if not pd.isna(df2.loc[index-1, "invest_cumu"]):
                df2.loc[index, "invest_cumu"] = (df2.loc[index, "invest_credibility"]+df2.loc[index-1, "invest_cumu"])/2
            else:
                df2.loc[index, "invest_cumu"] = 1-(abs(row["invest_score"] - row["StockScore"]))/2            
    else:
        df2.loc[index,"invest_score"] = None
        df2.loc[index,"invest_credibility"] = None
        if index == 0:
            df2.loc[index, "invest_cumu"] = None
        else:
            if not pd.isna(df2.loc[index-1, "invest_cumu"]):
                df2.loc[index, "invest_cumu"] = df2.loc[index-1, "invest_cumu"]
            else:
                df2.loc[index, "invest_cumu"] = None

        
    if cnbc_sentiment_count > 0:
        df2.loc[index,"cnbc_score"] = cnbc_sentiment_score/cnbc_sentiment_count
        df2.loc[index,"cnbc_credibility"] = 1-(abs(row["cnbc_score"] - row["StockScore"]))/2
        if index == 0:
            df2.loc[index,"cnbc_cumu"] = 1-(abs(row["cnbc_score"] - row["StockScore"]))/2
        else:
            if not pd.isna(df2.loc[index-1, "cnbc_cumu"]):
                df2.loc[index, "cnbc_cumu"] = (df2.loc[index, "cnbc_credibility"]+df2.loc[index-1, "cnbc_cumu"])/2
            else:
                df2.loc[index, "cnbc_cumu"] = 1-(abs(row["cnbc_score"] - row["StockScore"]))/2            
    else:
        df2.loc[index,"cnbc_score"] = None
        df2.loc[index,"cnbc_credibility"] = None
        if index == 0:
            df2.loc[index, "cnbc_cumu"] = None
        else:
            if not pd.isna(df2.loc[index-1, "cnbc_cumu"]):
                df2.loc[index, "cnbc_cumu"] = df2.loc[index-1, "cnbc_cumu"]
            else:
                df2.loc[index, "cnbc_cumu"] = None

            
    if market_sentiment_count > 0:
        df2.loc[index,"market_score"] = market_sentiment_score/market_sentiment_count
        df2.loc[index,"market_credibility"] = 1-(abs(row["market_score"] - row["StockScore"]))/2
        if index == 0:
            df2.loc[index,"market_cumu"] = 1-(abs(row["market_score"] - row["StockScore"]))/2
        else:
            if not pd.isna(df2.loc[index-1, "market_cumu"]):
                df2.loc[index, "market_cumu"] = (df2.loc[index, "market_credibility"]+df2.loc[index-1, "market_cumu"])/2
            else:
                df2.loc[index, "market_cumu"] = 1-(abs(row["market_score"] - row["StockScore"]))/2            
    else:
        df2.loc[index,"market_score"] = None
        df2.loc[index,"market_credibility"] = None
        if index == 0:
            df2.loc[index, "market_cumu"] = None
        else:
            if not pd.isna(df2.loc[index-1, "market_cumu"]):
                df2.loc[index, "market_cumu"] = df2.loc[index-1, "market_cumu"]
            else:
                df2.loc[index, "market_cumu"] = None

        
```
